```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Data

```python
# read data
df = pd.read_csv('data.csv')
df.head()
```

# Imputation

```python
df.isna().sum()
```

There are no missing values in the dataset. We can proceed with our analysis.

# Univariate Analysis

```python
plt.figure(figsize=(8, 6))

sc1 = plt.scatter(df['Socioeconomic Score'], df['Grades'], color='blue', label='Socioeconomic Score')
sc2 = plt.scatter(df['Study Hours'], df['Grades'], color='green', label='Study Hours')
sc3 = plt.scatter(df['Sleep Hours'], df['Grades'], color='red', label='Sleep Hours')
sc1 = plt.scatter(df['Attendance (%)'], df['Grades'], color='purple', label='Attendance (%)')

plt.title("Each Feature vs Grades")
plt.xlabel("Feature")
plt.ylabel("Grades")

plt.legend(loc='upper right')

plt.grid(True)
plt.show()
```

Let's analyze each feature one at a time:

- **Socioeconomic Score**: The points shoot up in a straight vertical line, indicating little to no correlation with Grades.

- **Study Hours**: The points rapidly increase but at an angle towards the right. This shows a positive correlation with Grades.

- **Sleep Hours**: The points are tightly packed with no apparent trend or strong correlation with Grades.

- **Attendance (%)**: The points are more sparse with no obvious linear correlation with Grades. However, it is possible that higher Attendance (%) contributes to higher Grades. More analysis needs to be done.


# Bivariate Analysis

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Socioeconomic Score'], df['Grades'], label='Socioeconomic Score', color='blue')
plt.scatter(df['Study Hours'], df['Grades'], label='Study Hours', color='green')

plt.xlabel('Socioeconomic Score and Study Hours', fontsize=12)
plt.ylabel('Grades', fontsize=12)
plt.title('Socioeconomic Score and Study Hours vs Grades', fontsize=14)

plt.legend(loc='lower right')

plt.grid(True)
plt.show()
```

We can see that there is a **weak positive correlation** between Socioeconomic Score and Grades. The points mostly cluster around 50 regardless of socioeconomic score, but there are some spread at higher scores. 

We can also see that there is a **strong positive correlation** between Study Hours and Grades. 

Since Study Hours has a much stronger relationship with grades, we will most likely get more accurate predictions if we use Study Hours as a feature in our model as opposed to Socioeconomic Score.
