```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Data

```python
# read data
df = pd.read_csv('data.csv')
df.head()
```

# Univariate Analysis

```python
plt.figure(figsize=(12, 8))

sc1 = plt.scatter(df['Socioeconomic Score'], df['Grades'], color='blue', label='Socioeconomic Score')
sc2 = plt.scatter(df['Study Hours'], df['Grades'], color='green', label='Study Hours')
sc3 = plt.scatter(df['Sleep Hours'], df['Grades'], color='red', label='Sleep Hours')
sc1 = plt.scatter(df['Attendance (%)'], df['Grades'], color='purple', label='Attendance (%)')

plt.title("Each Feature vs Grades")
plt.xlabel("Feature")
plt.ylabel("Grades")

plt.legend(loc='upper right')

plt.show()
```

From the graph, we can see that Socioeconomic Score, Study Hours, and Sleep Hours all have an overall **strong positive correlation** with Grades. Attendance(%) seems to also have a positive correlation with Grades but weaker than that of the other features.
